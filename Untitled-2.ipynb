```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Replace with your actual file path
df = pd.read_csv(r"C:\summer project\updated_crop_disease_fertilizer.csv")

# Show first few rows
print(df.head())

```

```python
df['Crop'] = df['crop_name'].str.strip().str.lower()
df['Disease'] = df['disease_name'].str.strip().str.lower()
```

```python
def recommend_fertilizer(crop, disease, data=df):
    crop = crop.strip().lower()
    disease = disease.strip().lower()
    
    result = data[(data['Crop'] == crop) & (data['Disease'] == disease)]
    
    if not result.empty:
        fertilizer = result['fertilizer_name'].values[0]
        quantity = result['quantity_to_use'].values[0]
        return f"Recommended Fertilizer: {fertilizer}, Quantity: {quantity}"
    else:
        return "No recommendation available for this crop-disease combination."
```

```python
print(recommend_fertilizer('Tomato', 'bacterial spot'))
```
